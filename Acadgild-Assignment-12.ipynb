```python
import numpy as np 
import pandas as pd
from pandasql import sqldf 
```

```python
# Read the following data set:
# https://archive.ics.uci.edu/ml/machine-learning-databases/adult/
# Rename the columns as per the description from this file:
# https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names
```

```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None,
               names=['age','workclass','fnlwgt','education','education-num','martial-status','occupation','relationship',
                        'race','sex', 'capital-gain','capital-loss','hours-per-week', 'native-country','class'])
```

```python
sqladb=df
```

```python
# Q1 Create a sql db from adult dataset and name it sqladb
# 1. Select 10 records from the adult sqladb
```

```python
# since it is very tedious to pass local and global parameter in sqldf function we will create
# a funtion that will take the query and return it with either local or global
pysql=lambda q:sqldf(q,globals())
```

```python
pysql("select * from sqladb LIMIt 10")
```

```python
# Q2 Show me the average hours per week of all men who are working in private sector
```

```python
pysql("select workclass,avg([hours-per-week]) from sqladb group by workclass having workclass = 'Private'")
```

```python
# Q3 Show me the frequency table for education, occupation and relationship, separately
```

```python
pysql("select education, count(education) from sqladb group by education")
```

```python
pysql("select occupation, count(occupation) from sqladb group by occupation")
```

```python
pysql("select relationship, count(relationship) from sqladb group by relationship")
```

```python
# Q4 Are there any people who are married, working in private sector and having a masters
# degree
```

```python
pysql("select * from sqladb where relationship ='Wife' and workclass='Private' and education='Masters'")
#since the result set is null that mean  no one is satisfying the above criteria
```

```python
#Q5. What is the average, minimum and maximum age group for people working in
# different sectors

pysql("select occupation,avg(age),max(age),min(age) from sqladb group by occupation")
```

```python
# 6. Calculate age distribution by country
# By using Derived Table

pysql("select Age_interval,count(*)  as Total from(select case when age<=25 then '1-25'"
      "when age between 26  and 50 then '26-50'"
      "else 'above 50' end as 'Age_interval' from sqladb)"
      "group by Age_interval")

#it can also be done using catch all bucket
```

```python
# 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and
# 'capital-loss
```

```python
pysql("select 'capital-gain' - 'capital-loss' as 'Net-Capital-Gain' from sqladb Limit 10")
#this should not come 0 for every row there is some issue with how the sysntax is being recognised by pandasql 
```
